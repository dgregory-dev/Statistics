```python
# Import necessary libraries
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters for normal distribution
mean_duration = 8  # mean in minutes
std_dev_duration = 2  # standard deviation in minutes

# Probability calculations
# a) Less than 5 minutes
prob_less_than_5 = stats.norm.cdf(5, mean_duration, std_dev_duration)

# b) Between 7 and 10 minutes
prob_between_7_and_10 = stats.norm.cdf(10, mean_duration, std_dev_duration) - stats.norm.cdf(7, mean_duration, std_dev_duration)

# c) More than 12 minutes
prob_more_than_12 = 1 - stats.norm.cdf(12, mean_duration, std_dev_duration)

# Print the results
print(f"Probability of a call lasting less than 5 minutes: {prob_less_than_5:.4f}")
print(f"Probability of a call lasting between 7 and 10 minutes: {prob_between_7_and_10:.4f}")
print(f"Probability of a call lasting more than 12 minutes: {prob_more_than_12:.4f}")

# Visualization of the normal distribution
x = np.linspace(0, 16, 1000)
pdf = stats.norm.pdf(x, mean_duration, std_dev_duration)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution PDF', color='blue')

# Shade areas for probabilities
plt.fill_between(x, pdf, where=(x < 5), color='red', alpha=0.5, label='P(X < 5)')
plt.fill_between(x, pdf, where=(x >= 7) & (x <= 10), color='green', alpha=0.5, label='P(7 < X < 10)')
plt.fill_between(x, pdf, where=(x > 12), color='orange', alpha=0.5, label='P(X > 12)')

plt.title('Normal Distribution of Customer Service Call Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()


```

```python
# Parameters for binomial distribution
n_customers = 20  # number of trials
p_satisfied = 0.85  # probability of success

# Probability calculations
# a) Exactly 15 satisfied customers
prob_exactly_15 = stats.binom.pmf(15, n_customers, p_satisfied)

# b) At least 18 satisfied customers
prob_at_least_18 = sum(stats.binom.pmf(k, n_customers, p_satisfied) for k in range(18, n_customers + 1))

# c) No more than 10 satisfied customers
prob_no_more_than_10 = stats.binom.cdf(10, n_customers, p_satisfied)

# Print the results
print(f"Probability of exactly 15 satisfied customers: {prob_exactly_15:.4f}")
print(f"Probability of at least 18 satisfied customers: {prob_at_least_18:.4f}")
print(f"Probability of no more than 10 satisfied customers: {prob_no_more_than_10:.4f}")

# Visualization of the binomial distribution
x = np.arange(0, n_customers + 1)
binom_pmf = stats.binom.pmf(x, n_customers, p_satisfied)

plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='skyblue', alpha=0.7, label='Binomial PMF')

# Highlight specific probabilities
plt.bar(15, prob_exactly_15, color='red', alpha=0.7, label='P(X = 15)')
plt.bar(np.arange(18, n_customers + 1), binom_pmf[18:], color='green', alpha=0.7, label='P(X >= 18)')
plt.bar(np.arange(0, 11), binom_pmf[:11], color='orange', alpha=0.7, label='P(X <= 10)')

plt.title('Binomial Distribution of Customer Satisfaction')
plt.xlabel('Number of Satisfied Customers')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid()
plt.show()
```

```python
# Parameters for Poisson distribution
lambda_calls = 30  # average calls per hour

# Probability calculations
# a) Exactly 25 calls
prob_exactly_25 = stats.poisson.pmf(25, lambda_calls)

# b) More than 35 calls
prob_more_than_35 = 1 - stats.poisson.cdf(35, lambda_calls)

# c) Between 20 and 40 calls
prob_between_20_and_40 = stats.poisson.cdf(40, lambda_calls) - stats.poisson.cdf(19, lambda_calls)

# Print the results
print(f"Probability of receiving exactly 25 calls: {prob_exactly_25:.4f}")
print(f"Probability of receiving more than 35 calls: {prob_more_than_35:.4f}")
print(f"Probability of receiving between 20 and 40 calls: {prob_between_20_and_40:.4f}")

# Visualization of the Poisson distribution
x = np.arange(0, 70)
poisson_pmf = stats.poisson.pmf(x, lambda_calls)

plt.figure(figsize=(10, 6))
plt.plot(x, poisson_pmf, marker='o', linestyle='-', color='blue', label='Poisson PMF')

# Highlight specific probabilities
plt.scatter(25, prob_exactly_25, color='red', label='P(X = 25)')
plt.fill_between(x, 0, poisson_pmf, where=(x > 35), color='green', alpha=0.5, label='P(X > 35)')
plt.fill_between(x, 0, poisson_pmf, where=((x >= 20) & (x <= 40)), color='orange', alpha=0.5, label='P(20 <= X <= 40)')

plt.title('Poisson Distribution of Call Centre Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.show()
```
